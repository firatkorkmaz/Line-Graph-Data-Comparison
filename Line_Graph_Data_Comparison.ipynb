```python
# Data Comparison Graph #

import random
import numpy as np
import statistics as st
import matplotlib.pyplot as plt

# Graph Settings
plt.figure(figsize=(20, 15))        # Size of the Graph
plt.rc('legend', fontsize=24)       # Font Size of the Legend Info ("Sample-1", "Sample-2")
plt.rc('lines', linewidth=6)        # Thickness of the Coloured Lines on the Graph
#plt.ylim([0, 1])                   # Limit of the Visible y-Axis (Optional)
#plt.xlim([10, 100])                # Limit of the Visible x-Axis (Optional)

################################################

# This Part will Take Your Own Data in List Type (with the Same Length!)
SAMP1 = []
SAMP2 = []

# Remove or Comment out This Part when You Assign Your Own Data to the SAMP1 & SAMP2 Lists
for i in range(1, 101):
    SAMP1.append(random.randint(1, 100) / (i * 10))
    SAMP2.append(abs(100 - random.randint(1, 100)) / (i * 10))

################################################
    
epoch = list(np.arange(0, len(SAMP1), 1))

# Plotting Data in the Graph
plt.plot(epoch, SAMP1, label="Sample-1", color="blue")    # Data Line-1 
plt.plot(epoch, SAMP2, label="Sample-2", color="red")     # Data Line-2

# Printing Additional Specific Values from the Data Before Plotting the Graph (Optional)
x = -1
print(f"First Values:      Last Values:")
print(f"Sample-1: {SAMP1[0]:.3f} || Sample-1: {SAMP1[x]:.3f}")
print(f"Sample-2: {SAMP2[0]:.3f} || Sample-2: {SAMP2[x]:.3f}")

################################################

# Graph Settings
plt.title("Comparison Graph\n", fontsize=28)             # Main Title of the Graph
plt.ylabel("y-Values\n", fontsize=26)                    # Name of the y-Axis
plt.xlabel("\nx-Values\n", fontsize=24)                  # Name of the x-Axis
plt.tick_params(axis='y', labelleft=True, labelright=True, labelsize=20)  # Label Size & Label Location
plt.tick_params(axis='x', labeltop=True, labelbottom=True, labelsize=20)  # Label Size & Label Location
plt.locator_params(axis='y', nbins=20)                   # Limiting Tick Numbers to Be Visible (y-Axis)
plt.locator_params(axis='x', nbins=20)                   # Limiting Tick Numbers to Be Visible (x-Axis)
#plt.xticks(epoch)                                       # This Makes Every Tick on the x-Axis Visible
plt.legend()
plt.show()
```
